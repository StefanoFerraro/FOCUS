```python
# check if dirs in public/projects/mazpie/offline* contain pretrained_models dir
import os
import sys
from termcolor import colored

def check_dir(dir):
    if not os.path.isdir(f"{dir}/pretrained_models"):
        return False
    return True

main_path = "/mnt/public/projects/mazpie"

for dir in [f for f in os.scandir(main_path) if f.is_dir()]: 
    if "offline_" in dir.name:
        if check_dir(dir.path):
            print(f"{dir.path}: OK")
        else:
            print(f"{dir.path}: missing pretrained_models")
    
```
