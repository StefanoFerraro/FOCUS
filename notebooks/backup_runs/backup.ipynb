```python
# from rclone_python import rclone
import rclone
import os
from pathlib import Path

with open("rclone.conf") as f:
   cfg = f.read()

result = rclone.with_config(cfg)
```

```python
main_path = Path("/mnt/public/projects/mazpie/")
projects = [
            "offline_reacher_benchmark_6",
            ]
```

```python
# zip pretrained models, and transfer to nextcloud via rclone
import shutil
import tqdm
import multiprocessing
        
def worker(project, dir):
    shutil.make_archive(project, 'zip', dir)
    rclone.with_config(cfg).copy(f"{project}.zip", "webdav:FOCUS++/pretrained_models")
    
jobs = []
for project in tqdm.tqdm(projects):
    dir = os.path.join(main_path, project, "pretrained_models")
    p = multiprocessing.Process(target=worker, args=(project, dir))
    jobs.append(p)
    p.start()

    # rclone.zip_and_transfer(dir, "nextcloud", "pretrained_models", "mazpie")
    
```

```python
for job in jobs:
    job.join()
```
